```python
import nltk
from nltk import pos_tag
#nltk.download('averaged_perceptron_tagger')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from string import punctuation
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
import numpy as np
from tqdm import tqdm
```

```python
stops = stopwords.words('english')
```

```python
def preproces_and_pos_tag(text):
    tokens = word_tokenize(text)
    tokens = [w for w in tokens if w.isalpha() and w.lower() not in stops]
    tokens_pos = pos_tag(tokens)
    return tokens_pos
```

```python
noun_tags = set(['NN', 'NNS', 'NNP', 'NNPS'])
def unigram_noun(tokens_pos):
    nouns = [w for w, pos in tokens_pos if pos in noun_tags]
    return ' '.join(nouns)
```

```python
def bigram_noun(tokens_pos):
    nouns_bi = []
    for i in range(1, len(tokens_pos)):
        pos_tags = set([tokens_pos[i - 1][1], tokens_pos[i][1]])
        if len(pos_tags & noun_tags) != 0:
            nb = tokens_pos[i - 1][0] + ' ' + tokens_pos[i][0]
            nouns_bi.append(nb)
    return ' '.join(nouns_bi) # да, тут будут повторы, но по-другому TfidfVectorizer не умеет
```

```python
%%time

all_df['tokens_pos'] = all_df['reviewText'].apply(preproces_and_pos_tag)
```

```python
%%time

all_df['unigrams'] = all_df['tokens_pos'].apply(unigram_noun)
all_df['bigrams'] = all_df['tokens_pos'].apply(bigram_noun)
```

```python
all_df[['unigrams', 'bigrams']]
```

```python
tf_idf1 = TfidfVectorizer()
uni_tfidf = tf_idf1.fit_transform(all_df['unigrams'])
```

```python
tf_idf2 = TfidfVectorizer(ngram_range=(2, 2))
bi_tfidf = tf_idf2.fit_transform(all_df['bigrams'])
```

```python
num_docs = all_df.shape[0]

def sort_indexes(matrix, vectorizer):
    ## сортировка спарс матриц :)))
    words = np.array(vectorizer.get_feature_names())
    rows, cols = matrix.nonzero()
    values = matrix.data
    
    entities = []
    for i in tqdm(range(num_docs)):
        relevant = (rows == i)
        rel_cols = cols[(relevant * cols).nonzero()[0]]
        rel_val = values[(relevant * values).nonzero()[0]]
        cols_values = zip(rel_cols, rel_val)
        sorted_cols_value = sorted(cols_values, key=lambda v: v[1], reverse=True)[:4]
        sorted_indexes = [i[0] for i in sorted_cols_value]
        entities.extend(words[sorted_indexes].tolist())
    return entities
```

## 3. Выделение n-грамм

Если я правильно поняла формулировку задания, то требуется выделить биграммы. Я выделяю биграммы для однословных и двусловных сущностей отдельно, и для обоих случаев смотрю оба контекста.

```python
ents_set = set(ents)
```

```python
def bigram_ne1(tokens_pos):
    # оставляю pos, чтобы переиспользовать имеющуюся колонку
    n_tokens = len(tokens_pos)
    ne_bi = []
    for i in range(1, n_tokens):
        w1, w2 = tokens_pos[i - 1][0], tokens_pos[i][0]
        if w1 in ents_set or w2 in ents_set:
            ne_bi.append((w1, w2))
    return ne_bi
```

```python
def bigram_ne2(tokens_pos):
    # оставляю pos, чтобы переиспользовать имеющуюся колонку
    n_tokens = len(tokens_pos)
    ne_bi = []
    for i in range(3, n_tokens):
        w1r, w2r, w1l, w2l = tokens_pos[i - 3][0], tokens_pos[i - 2][0], tokens_pos[i - 1][0], tokens_pos[i][0]
        wr = ' '.join([w1r, w2r])
        wl = ' '.join([w1l, w2l])
        if wr in ents_set or wl in ents_set:
            ne_bi.append((wr, wl))
    return ne_bi
```

```python
all_df['bigrams_NE1'] = all_df['tokens_pos'].apply(bigram_ne1)
all_df['bigrams_NE2'] = all_df['tokens_pos'].apply(bigram_ne2)
```

```python
ne1_bigrams = [ngram for doc in all_df['bigrams_NE1'].tolist() for ngram in doc]
ne2_bigrams = [ngram for doc in all_df['bigrams_NE2'].tolist() for ngram in doc]
```

```python
unigrams_all = []
for doc in all_df['tokens_pos'].tolist():
    tokens = [w[0] for w in doc]
    unigrams_all.extend(tokens)
```

```python
bigrams_all = []
for i in range(1, len(unigrams_all)):
    bigrams_all.append(' '.join([unigrams_all[i-1], unigrams_all[i]]))
```

```python
ne1_count = Counter(ne1_bigrams)
ne2_count = Counter(ne2_bigrams)
uni_count = Counter(unigrams_all)
bi_count = Counter(bigrams_all)
```

## 4. Ранжирование метриками

Частотный фильтр -- 5.

```python
from nltk.collocations import BigramAssocMeasures

metrics = BigramAssocMeasures()
```

```python
def get_marginals(ne_count, all_count):
    corpus_size = sum(ne_count.values())
    marginals = dict()
    already_seen = dict()
    for key, value in tqdm(ne_count.items()):
        joint = value
        r, l = key
        r_sum = all_count[r]
        l_sum = all_count[l]
        #if r_sum != 0 and l_sum != 0:
        marginals[key] = (joint, (r_sum, l_sum), corpus_size)
    return marginals
```

```python
ne1_count_short = {k: v for k, v in ne1_count.items() if v > 5}
ne2_count_short = {k: v for k, v in ne2_count.items() if v > 5}
```

```python
marginals1 = get_marginals(ne1_count_short, uni_count)
```

```python
marginals2 = get_marginals(ne2_count_short, bi_count)
```

```python
# Pointwise mutual information (PMI)
pmi1 = {k: metrics.pmi(*v) for k, v in marginals1.items()}
pmi2 = {k: metrics.pmi(*v) for k, v in marginals2.items()}
```
